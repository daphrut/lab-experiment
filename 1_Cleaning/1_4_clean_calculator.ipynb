# Cleaning 1.4 - Calculator cleaning

This notebook does the following:
    (1) Strips the calculator variables to numbers

## Set-up

```python
# Set-up
import pandas as pd
import numpy as np
import sys
from pathlib import Path
CODE_ROOT = Path.cwd().parents[0]
sys.path.append(str(CODE_ROOT))
import config
from openpyxl import Workbook
from openpyxl.styles import Font, Alignment
import os
```

```python
# Load data
labs = pd.read_csv(
    config.PROCESSED_DATA / "individual_processed_3.csv",
    keep_default_na=False,  # Keep "None" as a string, not NaN
    na_values=[""] # Only treat empty strings as NaN
)
```

## (1) Strip calculator variables to numbers

```python
# For all calculator variables, remove " kWh" and " tCO2e" suffixes
for var in ["energy", "co2"]:
    for e in ["total", "fridge", "freezer", "ult", "glassware", "microbio", "cryostat", "bath", "incubator", "heater", "it"]:
        energy_col = f"calc_{e}_energy"
        co2_col = f"calc_{e}_co2"
        labs[energy_col] = labs[energy_col].str.replace(" kWh", "", regex=False)
        labs[co2_col] = labs[co2_col].str.replace(" tCO2e", "", regex=False)
```

## Save processed dataset

```python
# Save processed dataset
labs.to_csv(config.PROCESSED_DATA / "individual_processed_4.csv", index = False)
```
