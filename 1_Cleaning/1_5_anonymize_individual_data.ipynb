## Set-up

```python
# Set-up
import pandas as pd
import numpy as np
import sys
from pathlib import Path
CODE_ROOT = Path.cwd().parents[0]
sys.path.append(str(CODE_ROOT))
import config
from openpyxl import Workbook
from openpyxl.styles import Font, Alignment
import os
```

```python
# Load data
labs = pd.read_csv(
    config.PROCESSED_DATA / "individual_processed_4.csv",
    keep_default_na=False,  # Keep "None" as a string, not NaN
    na_values=[""] # Only treat empty strings as NaN
)
```

```python
# Identify sensitive cols
rooms_cols = [c for c in labs.columns if c.startswith("rooms")]
share_equip_groups_cols = [c for c in labs.columns if c.startswith("share_equip_groups")]
share_space_groups_cols = [c for c in labs.columns if c.startswith("share_space_groups")]
comm_group_cols = [c for c in labs.columns if c.startswith("comm_group")]


sensitive_cols = ["institute_old", 
                  "Professor", 
                  "Email", 
                  "Source", 
                  "enum_firstname",
                  "enum_lastname",
                  "enum_email",
                  "foldername",
                  "Contact person (if different)",
                  "Contact email (if different)",
                  "Comments?",
                  "institute",
                  "Lab Group"
] + rooms_cols + share_equip_groups_cols + share_space_groups_cols + comm_group_cols
```

```python
# Separate df into anonymized and non-anonymized parts
non_anonymized_cols = ["labgroupid"] + sensitive_cols
labs_non_anonymized = labs[non_anonymized_cols].copy()
labs_anonymized = labs.drop(columns=sensitive_cols).copy()
```

## Save processed datasets

```python
# Save anonymized dataset
labs_anonymized.to_csv(config.PROCESSED_DATA / "individual_processed_5.csv", index = False)
```

```python
# Save sensitive dataset
labs_non_anonymized.to_csv(config.SENSITIVE_DATA / "sensitive_data.csv", index = False)
```
